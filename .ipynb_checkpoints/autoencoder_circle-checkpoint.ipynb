# Given a list of circle points, can we compress this information by representing these points with only 3-numbers --- circle centroid and the radius?

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

import sklearn
import tensorflow as tf
from tensorflow import keras

np.random.seed(42)
tf.random.set_seed(42)
```

# Generate circle points
The theta of each point is constant because we do not want the encoder to try to encode the thetas in the embedding.

```python
import random
import math

def getCirclePoints(center, radius, N=10):
    points = np.zeros([N,2])
    for i in range(N):
        theta = ((i*1.0)/N)*2*math.pi #random.uniform(0,2*math.pi)
        points[i,0] = center[0]+radius*math.cos(theta)
        points[i,1] = center[1]+radius*math.sin(theta)
    return points

points = getCirclePoints([10,20], 5, N=10)

fig = plt.gcf()
ax = plt.gca().axes
fig.set_size_inches(5, 5)
plt.scatter(points[:,0], points[:,1])
```

```python
params = []
X = []
for i in range(1500):
    center = [random.uniform(0,1), random.uniform(0,1)]
    radius = random.uniform(0,1)
    points = getCirclePoints(center, radius, N=10)
    params.append([center[0], center[1], radius])
    X.append(points)
    #X.append(points.flatten())
    
X = np.array(X)
params = np.array(params)
print(X.shape)
print(params.shape)

X_train = X[:1000,:,:]
X_test = X[1000:,:,:]

y_train = params[:1000,:]
y_test = params[1000:,:]

print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)


fig = plt.gcf()
ax = plt.gca().axes
fig.set_size_inches(5, 5)
plt.scatter(X_train[0][:,0], X_train[0][:,1])
# plt.scatter(trainX[0][::2], trainX[0][1::2])
```

# Train the encoder and the decoder

```python
tf.random.set_seed(42)
np.random.seed(42)

# default_activation = 'selu'
default_activation = 'linear'
# default_activation = 'relu'

stacked_encoder = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[X_train.shape[1], X_train.shape[2]]),
    keras.layers.Dense(20, activation=default_activation),
    keras.layers.Dense(3, activation=default_activation),
])
stacked_decoder = keras.models.Sequential([
    keras.layers.Dense(20, activation=default_activation, input_shape=[3]),
    keras.layers.Dense(X_train.shape[1]*X_train.shape[2], activation=default_activation),
    keras.layers.Reshape([X_train.shape[1], X_train.shape[2]])
])
stacked_ae = keras.models.Sequential([stacked_encoder, stacked_decoder])
stacked_ae.compile(loss="mean_squared_error",
                   batch_size=16,
                   #optimizer=keras.optimizers.SGD(lr=1.5), 
                   optimizer='adam',
                   metric='mse'
                  )
history = stacked_ae.fit(X_train, X_train, epochs=100,
                         validation_data=[X_test, X_test])
```

```python
plt.plot(np.arange(1,len(history.history['loss'])+1), history.history['loss'])
plt.plot(np.arange(1,len(history.history['loss'])+1), history.history['val_loss'])
# plt.title('Model Loss')
plt.ylabel('Loss', fontsize=14)
plt.xlabel('Epoch', fontsize=14)
# matplotlib.pyplot.xticks(x)
plt.legend(['Train', 'Test'], loc='upper right', fontsize=14)
plt.tight_layout()
plt.show()
```

# Autoencoder is able to reconstruct the circles using only length-3 encoding

```python
def show_reconstructions(model, X, N=5):
    reconstructions = model.predict(X[:N])
    fig = plt.figure(figsize=(N * 3, 6))
    for i in range(N):
        plt.subplot(2, N, 1 + i)
        plt.scatter(X[i][:,0], X[i][:,1])
        plt.subplot(2, N, 1 + N + i)
        plt.scatter(reconstructions[i][:,0], reconstructions[i][:,1])
        
# print('Training:')
# show_reconstructions(stacked_ae, X_train, N=5)
print('Testing:')
show_reconstructions(stacked_ae, X_test, N=5)
```

```python
y_pred_coding = stacked_encoder(X_test).numpy()
print(y_pred_coding.shape)
print(y_test.shape)
```

# Circle centroid and radius correlated with the estimated encoding

```python
param_labels = ['Circle x', 'Circle y', 'Radius']
fig = plt.figure(figsize=(y_pred_coding.shape[1]*4, y_test.shape[1]*4))
props = dict(boxstyle='round', facecolor='white', alpha=0.5)
for i in range(y_pred_coding.shape[1]):
    for j in range(y_test.shape[1]):
        plt.subplot(y_pred_coding.shape[1], y_test.shape[1], 1 + i*y_pred_coding.shape[1]+j)
        plt.scatter(y_pred_coding[:,i], y_test[:,j])
        plt.xlabel('Autoencoder coding: ' + str(i))
        plt.ylabel(param_labels[j])
        np.corrcoef(y_pred_coding[:,i], y_test[:,j])
        textstr = 'R={:.3f}'.format(np.corrcoef(y_pred_coding[:,i], y_test[:,j])[0,1])
        ax = plt.gca().axes
        ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=16, verticalalignment='top', bbox=props)
        
plt.tight_layout()
```
